```python
# Count words in the bible uppercase.txt using for loop
number_of_words = 0
with open(r'bible_uppercase.txt','r') as file:
    for line in file:
        words = line.split()
        for word in words:
            number_of_words += 1
print(number_of_words)
```

```python
# Count words in the bible uppercase.txt using map
def f(words):
    word_count = 0
    words = words.split()
    for word in words:
        word_count += 1        
    return word_count

with open('bible_uppercase.txt', 'r') as file:
    contents = file.read()
    lines = contents.split('\n')
    word_counts = map(f, lines)
    print(sum(list(word_counts)))
```

```python
# Count words in the bible uppercase.txt using a Pythonâ€™s multipro- cessing API.
def f(words):
    word_count = 0
    words = words.split()
    for word in words:
        word_count += 1        
    return word_count

from multiprocessing import Pool

if __name__ == '__main__':
  with open('bible_uppercase.txt', 'r') as file:
      contents = file.read()
      lines = contents.split('\n')
      with Pool(5) as p:
        word_counts = p.map(f, lines)
      print(sum(list(word_counts)))
```
